```python
# values_count return how many count and percentage
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

```python
# visualiazation of bar graphs
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts(),edgecolor = 'r',width = 0.7)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')

# Plotting the bar chart
plt.bar(df['is_canceled'].value_counts().index, df['is_canceled'].value_counts(),edgecolor='k', width=0.7)

plt.xlabel('Reservation Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1= sns.countplot(x= 'hotel',hue = 'is_canceled',data=df,palette = 'Blues')
legend_labels = ax1.get_legend_handles_labels()

plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python

# resort hotel in percentage we can see

resort_hotel = df[df['hotel']  == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# same we see city hotel percentage 
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# get extracte month column and after that visualization months wise
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x='month',hue = 'is_canceled',data = df,palette = 'bright')
legend_labels = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
plt.title('ADR PER MONTH', fontsize=30)

# Filter the DataFrame for canceled bookings (assuming 'is_canceled' == 1 represents canceled bookings)
canceled_bookings = df[df['is_canceled'] == 1]

# Group by month and calculate the sum of ADR
monthly_adr_sum = canceled_bookings.groupby('month')['adr'].sum().reset_index()

# Use seaborn's barplot function to plot the data
sns.barplot(x='month', y='adr', data=monthly_adr_sum)

plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

```python
# top 10 country visulization in pie chart.
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('TOP 10 COUNTRY WITH RESERVATION CANCELED')
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]

plt.figure(figsize=(8, 8))
plt.title('TOP 10 COUNTRY WITH RESERVATION CANCELED')

# Plotting the bar chart
plt.bar(top_10_country.index, top_10_country.values, color='skyblue', edgecolor='r')

# Setting labels for x-axis ticks
plt.xticks(rotation=45, ha='right')

plt.xlabel('Country')
plt.ylabel('Number of Cancellations')
plt.show()
```
